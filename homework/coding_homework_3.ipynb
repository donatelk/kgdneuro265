```python
#Import Packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Import data
v_rest_sst_url= 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest_sst= np.loadtxt(v_rest_sst_url)

v_rest_pvalb_url= 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_pvalb= np.loadtxt(v_rest_pvalb_url)
```

```python
#Find means
mean_sst= np.mean(v_rest_sst)
mean_pvalb= np.mean(v_rest_pvalb)

#t-test
test= stats.ttest_ind(v_rest_sst, v_rest_pvalb, equal_var= False)

#Conditional histogram
if test.pvalue < 0.05:
  fig, ax= plt.subplots(1,1)
  ax.hist(v_rest_sst, alpha = 0.5, label ='SST')
  ax.hist(v_rest_pvalb, alpha = 0.5, label ='Pvalb')
  ax.set_xlabel('Resting Membrane Potential (mV)')
  ax.set_ylabel('Frequency')
  plt.show()
```

```python
#Make a function
def vrestHist():
  """
  Makes a histogram of resting membrane potential values for sst and pvalb cells when their means are satistically different
  """
  #Find means
  mean_sst= np.mean(v_rest_sst)
  mean_pvalb= np.mean(v_rest_pvalb)

  #t-test
  t_test= stats.ttest_ind(v_rest_sst, v_rest_pvalb, equal_var= False)

  #Conditional histogram
  if t_test.pvalue < 0.05:
    fig, ax= plt.subplots(1,1)
    ax.hist(v_rest_sst, alpha = 0.5, label ='SST')
    ax.hist(v_rest_pvalb, alpha = 0.5, label ='Pvalb')
    ax.set_xlabel('Resting Membrane Potential (mV)')
    ax.set_ylabel('Frequency')
    plt.show()
```

```python
#Call the function
vrestHist()
```

```python
#Create loop for scatter plot

fig, ax = plt.subplots(1,1)
#Make the loop
for i in range(len(v_rest_sst)):
    x = v_rest_sst[i]
    y = v_rest_pvalb[i]

#Conditionals
    if (x > -60) and (y > -60):
      c = 'red'
    elif (x > -60) and (y < -60):
      c = 'green'
    elif (x < -60) and (y < -60):
      c = 'blue'
    elif (x < -60) and (y > -60):
      c = 'purple'
    ax.scatter(x, y, c = c)
#Labels and print
ax.set_xlabel('Resting Membrane Potential of Sst Cells')
ax.set_ylabel('Resting Membrane Potential of Pvalb Cells')
plt.show()
```
