```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
hits_url= 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_hits_hw5.csv'
fa_url= 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_fa_hw5.csv'

acc_hits= np.loadtxt(hits_url, delimiter=',')
acc_fa= np.loadtxt(fa_url, delimiter=',')
```

```python
#Shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#Create time variable
time= np.linspace(0,4,8001)
```

```python
#sampling rate
sampling_rate= 8001/4
sampling_rate
```

```python
#Plot of first trial of hits and fa
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(time, acc_hits[0])
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Voltage (mV)')
ax[0].set_title('Acc Hits Trial 1')

ax[1].plot(time, acc_fa[0])
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Voltage (mV)')
ax[1].set_title('Acc FA Trial 1')
```

There are about 4-5 visable peaks in a 1 second period at the end of hits trial one, so the dominant oscillation is around 4-5Hz. The first trial for fa seems to have about 2 peaks persecond, so the dominant oscillation would be about 2Hz.

```python
#Calculate and graph ERP of Acc Hits
plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time(s)')
plt.ylabel('Voltage (mV)')
plt.title('ERP of Acc Hits')

#vertical and horizontal lines
plt.plot([2,2], [-50,50], 'k', lw=2)
plt.hlines(0, time[0], time[-1], color= 'black')
```

Yes, there is an increase in activity following the screen touch, with the voltage droping then rising again.

```python
#Calculate and graph ERP of Acc Hits and FAs
plt.plot(time, acc_hits.mean(0))
plt.plot(time, acc_fa.mean(0))
plt.xlabel('Time(s)')
plt.ylabel('Voltage (mV)')
plt.title('ERP of Acc Hits and FAs')

#vertical and horizontal lines
plt.plot([2,2], [-50,50], 'k', lw=2)
plt.hlines(0, time[0], time[-1], color= 'black')
```

This graph shows that neural activity reacts more to the screen touch than the false alarms, with little change in voltage following the screen touch for false alarm ERP.
